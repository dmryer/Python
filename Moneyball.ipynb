```python
# Create a simple barchart
# https://pythonspot.com/en/matplotlib-bar-chart/
# Matplotlib Bar chart

import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

% matplotlib inline
 
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]
 
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming Language Usage')
 
plt.show()
```

```python
# Create a barchart for MLB data

import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

# Plot the figure in the webpage and define size 
% matplotlib inline
width = 12
height = 10
plt.figure(figsize=(width, height))
 
objects = ('Marlins','Athletics','Nationals/Expos','Pirates','Rays','Twins','Royals','Padres','Reds','Brewers',
           'Indians','Astros','White Sox','Blue Jays','Diamondbacks','Rockies','Cardinals','Phillies','Tigers',
           'Rangers','Giants','Braves','Mariners','Angels','Orioles','Cubs','Mets','Dodgers','Red Sox','Yankees')
y_pos = np.arange(len(objects))
wins = [2.27,1.80,1.79,1.76,1.74,1.69,1.54,1.40,1.35,1.33,1.28,1.26,1.24,1.23,1.15,1.14,1.07,1.07,1.06,1.03,
1.02,1.01,1.00,0.97,0.96,0.90,0.85,0.80,0.76,0.63]
 
barlist=plt.barh(y_pos, wins, align='center', alpha=0.5)
barlist[1].set_color('g') # set the second bar to green for the Oakland A's
#plt.barh(y_pos, wins, align='center', alpha=0.5)

plt.yticks(y_pos, objects)
plt.xlabel('Wins/$M',fontweight="bold")
plt.ylabel ('Major League Baseball Team',fontweight="bold")
plt.title('Average Wins/$M (1998-2016)',fontweight="bold")

#plt.axis('tight')
plt.margins(0.01, 0.01)  # if you'd prefer a bit of space, use plt.margins(0.05, 0) to pad the limits 
plt.show()
```
