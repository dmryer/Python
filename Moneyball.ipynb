# Moneyball
- Summary analysis evaluating effectiveness of methodology during the 19 years.
- Intial examination manually evalutes aggregate Major League Baseball (MLB) data already processed in a spreadsheet.

```python
#%% Create a barchart for MLB data
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline

# Define figure size for inline presentation in webpage  
width = 8
height = 6
plt.figure(figsize=(width, height))

# Create data for the chart
objects = ('Marlins','Athletics','Nationals/Expos','Pirates','Rays','Twins','Royals','Padres','Reds','Brewers',
           'Indians','Astros','White Sox','Blue Jays','Diamondbacks','Rockies','Cardinals','Phillies','Tigers',
           'Rangers','Giants','Braves','Mariners','Angels','Orioles','Cubs','Mets','Dodgers','Red Sox','Yankees')
objects = np.flipud(objects)
y_pos = np.arange(len(objects))
wins = [2.27,1.80,1.79,1.76,1.74,1.69,1.54,1.40,1.35,1.33,1.28,1.26,1.24,1.23,1.15,1.14,1.07,1.07,1.06,1.03,
1.02,1.01,1.00,0.97,0.96,0.90,0.85,0.80,0.76,0.63]
wins = np.flipud(wins)

# Plot the figure 
barlist=plt.barh(y_pos, wins, align='center', alpha=0.5)
barlist[28].set_color('g') # set the second bar to green for the Oakland A's
plt.yticks(y_pos, objects)
plt.xlabel('Wins/$M',fontweight="bold")
plt.ylabel ('Major League Baseball Team',fontweight="bold")
plt.title('Average Wins/$M (1998-2016)',fontweight="bold")
plt.margins(0.01, 0.01)                    
plt.show()
```

- Billy Beane and Oakland Athletics Ranking over 19 Years:
- Goal: Purchase hits, that produce runs, that result in wins.  So who buys wins at the best price?
- Data over the last two decades since beginning in 1998 - his Team is 2nd best in Entire MLB League and 1st in American League 
- 1.5 standard deviations above average number of wins per $M

# Extracting MLB win data from excel file 

```python
#%% Data extraction from xlxs file
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
% matplotlib inline

# Read Win data from MLB Excel file
df = pd.read_excel('MLB.xlsx', sheetname = 'WinData')
meandf = df.mean(axis=1)
sortmeandf= meandf.sort_values(ascending=False)
ax = sortmeandf.plot(kind='bar', title ="Average Wins Per Season (1998-2016)", legend=False, fontsize=12)
ax.set_xlabel("Major League Baseball Team",fontsize=12, fontweight="bold")
ax.set_ylabel("Average Wins",fontsize=12, fontweight="bold")
plt.show()
```

- Top Seven teams that beat the A's are all premier big market teams (St. Louis not really big market)
- Almost one standard deviation above average number of wins per season

# Extracting MLB salary data from excel file 

```python
#%% Data extract from xlxs file, manipulation and visualization
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
% matplotlib inline

# read in Salary data from Major League Baseball excel file
df = pd.read_excel('MLB.xlsx', sheetname = 'SalaryData')
meandf = df.mean(axis=1)
sortmeandf= meandf.sort_values(ascending=False)/1000000
ax = sortmeandf.plot(kind='bar', title ="Average Salary ($M) Per Season (1998-2016)",legend=False, fontsize=12)
ax.set_xlabel("Major League Baseball Team",fontsize=12, fontweight="bold")
ax.set_ylabel("Average Salary ($M)",fontsize=12, fontweight="bold")
plt.show()
```

- Average team salary 27th of 30 
- Over one standard deviation below the average salary

# Scatter plot of MLB win and salary data 

```python
#%%  MLB Salary Data extract from xlxs file, manipulation and visualization
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
%matplotlib inline

# read in Win data from MLB excel file
df1 = pd.read_excel('MLB.xlsx', sheetname = 'WinData')
meandf1 = df1.mean(axis=1)
sortmeandf1= meandf1.sort_values(ascending=False)

# read in Salary data from MLB excel file
df2 = pd.read_excel('MLB.xlsx', sheetname = 'SalaryData')
meandf2 = df2.mean(axis=1)
sortmeandf2= meandf2.sort_values(ascending=False)/1000000

# Create scatterplot of Salary vs. Wins and add data labels
plt.figure(figsize=(20,8))
plt.scatter(sortmeandf2, sortmeandf1)
for i, txt in enumerate(sortmeandf1.index):
    plt.annotate(txt, (sortmeandf2[i],sortmeandf1[i]),fontsize=14)

plt.title("Wins/$M (1998-2016)")    
plt.xlabel("Salary $M")
plt.ylabel("Avg Wins/yr")
plt.tight_layout()
plt.show()
```

# Bar chart of MLB win and salary data

```python
#%%  MLB Salary Data extract from xlxs file, manipulation and visualization
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
%matplotlib inline

# read in Win data from excel sort both by index
df1 = pd.read_excel('MLB.xlsx', sheetname = 'WinData')
sortdf1= df1.sort_index(ascending=False)

# read in Salary data from excel sort both by index
df2 = pd.read_excel('MLB.xlsx', sheetname = 'SalaryData')
sortdf2= df2.sort_index(ascending=False)/1000000

# Divide dataframes to obtain Win/$M
sortdf12 = sortdf1/sortdf2
meansortdf12 = sortdf12.mean(axis=1)   # take average of all years
sortmeandsortf12= meansortdf12.sort_values(ascending=False) # sort avg by value

#Create bar chart of sorted averages 
sortmeandsortf12.plot(kind='bar', title ="Average #Wins Season per $M (1998-2016)",figsize=(8,6),legend=False, fontsize=12)
plt.title("Average Wins/$M (1998-2016)")    
plt.xlabel("Major League Baseball Team")
plt.ylabel("Avg Wins/yr/$M")
plt.tight_layout()
plt.show()
```

# A quick look at the 2016 Season

```python
#%%  MLB Salary Data extract from xlxs file, manipulation and visualization

import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt

%matplotlib inline

# read in Win data from excel sort both by index
df1 = pd.read_excel('MLB.xlsx', sheetname = 'WinData')
sortdf1= df1.sort_index(ascending=False)
Wins2016 = sortdf1[2016]

# read in Salary data from excel sort both by index
df2 = pd.read_excel('MLB.xlsx', sheetname = 'SalaryData')
sortdf2= df2.sort_index(ascending=False)/1000000
Salary2016 = sortdf2[2016]

# calculate Win/$M and sort in decreasing order
WinsSalary2016 = Wins2016/Salary2016
sortdf2016 = WinsSalary2016.sort_values(ascending=False) # sort by value

#Create bar chart
sortdf2016.plot(kind='bar', title ="Wins per $M (2016)",figsize=(8,6),legend=False, fontsize=12)
plt.title("Wins/$M (2016)")    
plt.xlabel("Major League Baseball Team")
plt.ylabel("Wins/$M")
plt.tight_layout()
plt.show()
```

- Even looking at the current season, there is a wide diparity between Win/$ metric
