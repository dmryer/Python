- Generate data, build histogram and fit a normal distribution

```python
# Fitting a Normal distribution to 1D data

import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
% matplotlib inline

# Generate some data for this demonstration.
data = norm.rvs(10.0, 2.5, size=500)

# Fit a normal distribution to the data:
mu, std = norm.fit(data)

# Plot the histogram.
plt.hist(data, bins=25, normed=True, alpha=0.6, color='b')

# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2, color='r')
title = "Normal Distribution: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)
plt.show()
```

- Similar exercise using numpy.random.normal

```python
# Generate data, build histogram and fit a normal distribution using numpy.random.normal

import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline

# mean and standard deviation
mu, sigma = 0, 0.1 
s = np.random.normal(mu, sigma, 1000)
abs(mu - np.mean(s)) < 0.01
abs(sigma - np.std(s, ddof=1)) < 0.01
 
count, bins, ignored = plt.hist(s, 30, normed=True, color='b')

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
    linewidth=2, color='r')
title = "Normal Distribution: mu = %.2f,  std = %.2f" % (mu, sigma)
plt.title(title)
plt.show()
```

- Using scipy.stats.uniform - display uniform pdf, generate data, build histogram 

```python
#%% Using scipy.stats.uniform - display uniform pdf, generate data, build histogram  

from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np

% matplotlib inline

fig, ax = plt.subplots(1, 1)

# Calculate a few first moments:
mean, var, skew, kurt = uniform.stats(moments='mvsk')

# Display the probability density function (pdf):
x = np.linspace(uniform.ppf(0.01),
                uniform.ppf(0.99), 100)
ax.plot(x, uniform.pdf(x),
        'r-', lw=5, alpha=0.6, label='uniform pdf')
        
# Freeze the distribution and display the frozen pdf:
rv = uniform()
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')

# Generate random numbers:
r = uniform.rvs(size=1000)

# And compare the histogram:
ax.hist(r, normed=True, alpha=0.2, color='b')
ax.legend(loc='best', frameon=False)
title = "Uniform Distribution: mu = %.2f,  var = %.2f" % (mean, var)
plt.title(title)
plt.show()
```

- Demonstrate the result of combining multiple uniform random distributions

```python
#%% Sums of uniform random values
from scipy.stats import uniform
import matplotlib.pyplot as plt

% matplotlib inline
    
# First Variable        
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, sharex='col', sharey='row')

# Generate random numbers:
r1 = uniform.rvs(size=1000)
# Plot histogram:
ax1.hist(r1, normed=True, histtype='stepfilled', alpha=0.2, color='g')

# Second Variable     
# Generate random numbers:
r2 = uniform.rvs(size=1000)
# Plot histogram:
ax2.hist(r2, normed=True, histtype='stepfilled', alpha=0.2, color='g')

# Sum of two variables 
rtwo = (r1 + r2)/2        
ax3.hist(rtwo, normed=True, histtype='stepfilled', alpha=0.2, color='r')
   
# Sum of three variables 
r3 = uniform.rvs(size=1000)
rthree = (r1 + r2 + r3)/3    
ax4.hist(rthree, normed=True, histtype='stepfilled', alpha=0.2, color='r')
   
# Sum of five variables 
r4 = uniform.rvs(size=1000)
r5 = uniform.rvs(size=1000)
rfive = (r1 + r2 + r3 + r4 + r5)/5    
ax5.hist(rfive, normed=True, histtype='stepfilled', alpha=0.2, color='b')

# Sum of ten variables 
r6 = uniform.rvs(size=1000)
r7 = uniform.rvs(size=1000)
r8 = uniform.rvs(size=1000)
r9 = uniform.rvs(size=1000)
r10 = uniform.rvs(size=1000)
rten = (r1 + r2 + r3 + r4 + r5 + r6 + r7 + r8 + r9 + r10)/10    
ax6.hist(rten, normed=True, histtype='stepfilled', alpha=0.2, color='b')  
plt.show() 
```
